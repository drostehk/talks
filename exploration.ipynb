# Exploration with Pandas

```python
%matplotlib inline
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib
matplotlib.style.use('ggplot')
```

### Setting up the Data

```python
ts = pd.Series(np.random.randn(100), index=pd.date_range('1/1/2000', periods=100))
ts = ts.cumsum()

df1 = pd.DataFrame(np.random.randn(100, 4), index=ts.index, columns=list('ABCD'))
df2 = pd.DataFrame({'a': np.random.randn(1000) + 1, 'b': np.random.randn(1000),'c': np.random.randn(1000) - 1}, columns=['a', 'b', 'c'])
df3 = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
df4 = pd.DataFrame(np.random.rand(50, 4), columns=['a', 'b', 'c', 'd'])
df5 = pd.DataFrame(np.random.randn(1000, 4), columns=['a', 'b', 'c', 'd'])
```

### Line Plot

```python
df1.head()
```

```python
df1.plot(figsize=(16,8));
```

### Histograms

```python
df2.head()
```

```python
df2.plot(kind='hist', alpha=0.5, figsize=(16,8));
```

### Area Plot

```python
df3.head()
```

```python
df3.plot(kind='area', stacked=False, figsize=(16,8));
```

### Scatter Plot

```python
df4.head()
```

```python
df4.plot(kind='scatter', x='a', y='b', s=df4['c']*200, figsize=(16,8));
```

### Scatter Matrix

```python
from pandas.tools.plotting import scatter_matrix

scatter_matrix(df5, alpha=0.6, figsize=(16, 16), diagonal='hist');
```
